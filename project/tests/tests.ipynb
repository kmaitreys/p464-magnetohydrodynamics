```julia
function gradient(arr, dr)
    n = length(arr)
    grad = similar(arr)
    grad[1] = (arr[2] - arr[1]) / dr
    grad[end] = (arr[end] - arr[end-1]) / dr
    for i in 2:length(arr)-1
        grad[i] = (arr[i+1] - arr[i-1]) / (2 * dr)
    end
    return grad
end


function get_laplacian(y::Vector{Float64}, dr)
    return gradient(gradient(y, dr), dr)
end

test = [1.0, 2.0, 4.0, 7.0, 11.0, 16.0]

println(gradient(test, 1.5))
```

```julia
using Plots

# Function to plot every 10th row of a 2D array on the same plot
function plot_every_10th_row(data::Matrix)
    num_rows = size(data, 1)
    p = plot(legend=:topleft)  # Initialize plot object
    for i in 10:10:num_rows
        plot!(p, data[i, :], label="Row $i")
    end
    xlabel!("Index")
    ylabel!("Value")
    title!("Plot of Every 10th Row")
    return p
end

# Example usage
data = rand(50, 10)  # Example 50x10 random matrix
plot_every_10th_row(data)
```
